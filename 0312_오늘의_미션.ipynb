=== 학생부 ===

이름: 봉봉 (첫글자: 봉)

나이: 29살

과목: ['파이썬', '수학', '영어']

학번: (2, 3, 15)

첫번째 과목: 파이썬

좌우명: 오늘도 열코딩!

```python
dic = {"이름": "봉봉", "나이": "29살", "과목": ["파이썬","수학","영어"] ,"학번":(2,3,15)}
dic["좌우명"] = "오늘도 열코딩!"
print("===  학생부 ===")
print("이름:",dic["이름"][0])
print("나이:",dic["나이"])
print("과목:",dic["과목"])
print("학번:",dic["학번"])
print("첫번째 과목:",dic["과목"][0])
print("좌우명:",dic["좌우명"])
```

```python
# ✅ 괄호 정리

# [] 대괄호 = 리스트 (순서 있는 목록)
list = ["파이썬", "수학", "영어"]

# {} 중괄호 = 딕셔너리 (키:값 쌍)
dic = {"이름": "봉봉", "나이": 29}

# () 소괄호 = 튜플 (수정 불가능한 목록)
tuple = (2, 3, 15)


# ✅ 따옴표 정리

# 글자(문자열) → 따옴표 O
{"이름": "봉봉"}

# 숫자 → 따옴표 X (계산 가능)
{"나이": 29}

# 숫자인데 따옴표 쓰면 → 문자열이 되어버림 (계산 불가)
{"나이": "29"}  # 29 + 1 하면 오류!


# ✅ 헷갈릴 때 이 질문!
# 키:값 쌍이 있어?  →  {} 딕셔너리
# 그냥 목록이야?   →  [] 리스트
# 수정하면 안돼?   →  () 튜플
# 글자야?          →  따옴표 O
# 숫자야?          →  따옴표 X
```
