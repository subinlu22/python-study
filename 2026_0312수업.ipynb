```python
# 점수가 90 이상이면 최우수
# 점수가 70 이상이면 우수
# 점수가 60 이상이면 합격
# 점수가 60 미만이면 불합격
score=65
if score <= 60:
  print("불합격")
elif score <= 60:
  print("합격")
elif score >= 70:
  print("합격")
else:
  print("최우수")

```
